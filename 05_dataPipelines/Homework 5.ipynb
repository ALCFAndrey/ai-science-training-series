```python
import pandas as pd
```

```python
file1 = open('output.txt', 'r')
Lines = file1.readlines()
```

```python
mytext = []
for line in Lines:
    if line[0] == "m" and line[1] == "e" and line[2] == "a" and line[3] == "n":
        mytext.append(line.strip())
```

```python
import re

means = []
std = []
for x in mytext:
    z = re.findall(r'\d+.\d*',x)
    means.append(z[0])
    std.append(z[1])
    
```

```python
parallel_threads = [128,128,128,128,64,64,64,64,32,32,32,32]
prefetch_buffer_size = [1,2,4,8]*3

print(len(parallel_threads))
print(len(prefetch_buffer_size))


df = pd.DataFrame(data={"parallel_threads":parallel_threads,"prefetch_buffer_size":prefetch_buffer_size,"mean":means,"std":std})
df
```
